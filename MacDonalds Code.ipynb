```python
mcdonalds['Like'] = mcdonalds["Like"].replace({'I hate it!-5':'-5','I love it!+5':'+5'})

sns.catplot(x='Like',y='Age', data=mcdonalds, height=5,aspect=2,palette='Set2',kind='swarm')

plt.title('Likelyness of Mcdonald w.r.t Age')
plt.show()
```

```python
#Converting 11 cols with yes = /no=0

mapping = {'Yes':1, 'No': 0}

columns_to_convert = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for col in columns_to_convert:
    mcdonalds[col] = mcdonalds[col].map(mapping)
    
mcdonalds
```

```python
#Histogram of the each attributes
plt.rcParams['figure.figsize'] = (12,14)
mcdonalds.hist()
plt.show()
```

```python
# considering only first 11 attributes 

st_11_col = mcdonalds.loc[:,columns_to_convert]
st_11_col
```

```python
x = mcdonalds.loc[:,columns_to_convert].values
x
```

```python
pca_data = preprocessing.scale(x)

pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)

names = ['pc1','pc2','pc3','pc4','pc5',
         'pc6','pc7','pc8','pc9','pc10','pc11']

pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# correlation coefficient between original variables and the component

loadings = pca.components_

num_pc = pca.n_features_
pc_list = ['PC'+str(i) for i in list(range(1,num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = st_11_col.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True)
plt.show()
```

```python
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
pca_scores = PCA().fit_transform(x)

cluster.biplot(cscore=pca_scores, loadings=loadings, labels=mcdonalds.columns.values,
              var1=round(pca.explained_variance_ratio_[0]*100, 2),var2=round(pca.explained_variance_ratio_[1]*100,2),show=True,dim=(10,5))
```

### EXTRACTING SEGMENTS



```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualize = KElbowVisualizer(model, k=(1,12)).fit(st_11_col)
visualize.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(st_11_col)

mcdonalds['cluster_num'] = kmeans.labels_
print(kmeans.labels_)
print(kmeans.inertia_)
print(kmeans.n_iter_)
print(kmeans.cluster_centers_)
```

```python
from collections import Counter
Counter(kmeans.labels_)
```

```python
sns.scatterplot(data=pf, x='pc1', y='pc2', hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='X', c='r', s=80, label='centroids')
plt.legend()
plt.show()
```

##### Selecting target segment

```python
#Calculating the mean
#Visit frequency
from sklearn.preprocessing import LabelEncoder
mcdonalds['VisitFrequency'] = LabelEncoder().fit_transform(mcdonalds['VisitFrequency'])
visit = mcdonalds.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
mcdonalds['Like'] = LabelEncoder().fit_transform(mcdonalds['Like'])
Like = mcdonalds.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
#Gender
mcdonalds['Gender'] = LabelEncoder().fit_transform(mcdonalds['Gender'])
Gender = mcdonalds.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on='cluster_num', how='left')
segment
```

```python
plt.figure(figsize = (9,4))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=segment,s=400, color="r")
plt.title("Simple segment evaluation plot for the fast food data set",
          fontsize = 15) 
plt.xlabel("Visit", fontsize = 12) 
plt.ylabel("Like", fontsize = 12) 
plt.show()
```
